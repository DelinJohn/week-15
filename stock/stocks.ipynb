```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('all_stocks_5yr.csv')
data.head()
```

```python
data.info()
```

```python
data['date']=pd.to_datetime(data['date'])
```

```python
data.info()
```

```python
data['year']=data['date'].dt.year
```

```python
data_year=data.groupby('year')[['open','high','low','close']].mean()
```

```python
for i in ['open','high','low','close']:
    sns.lineplot(x='year',y=i,data=data_year)
    plt.show()
```

```python
sns.scatterplot(data=data,x='close',y='open')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('GOOG.csv')
```

```python
data.tail()
```

```python
data.info()
```

```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
sns.histplot(x='Volume',data=data)
```

```python
sns.barplot(x=data['Date'].dt.year,y='Volume',data=data)
```

```python
data['Year']=data['Date'].dt.year
```

```python
by_year=data.groupby(['Year'])[['Open','High','Low','Close','Adj Close','Volume']].mean()
```

```python
for i in ['Open','High','Low','Close','Adj Close','Volume']:
    sns.lineplot(data=by_year,x='Year',y=i)
    plt.show()
```

```python
sns.scatterplot(data=data,y='Open',x='Close')
```

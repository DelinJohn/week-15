```python
import pandas as pd
import seaborn as sns
```

```python
customer_data=pd.read_csv('Customers.csv',encoding='latin')
customer_data
```

```python
products_data=pd.read_csv('Products.csv')
products_data.info()
```

```python
sales_data=pd.read_csv('Sales.csv')
sales_data
```

```python
data=pd.merge(left=customer_data,right=sales_data,on='CustomerKey')
data.head()
```

```python
data=pd.merge(left=data,right=products_data ,on='ProductKey')
```

```python
data.info()
```

```python
data['Birthday']=pd.to_datetime(data['Birthday'])
data['Age']=pd.to_datetime('today')-data["Birthday"]
data['Age']=data['Age'].dt.days/365
data['Age']
```

```python
data.drop(['State Code','Birthday'])
```
